```python
import numpy as np
import matplotlib
import pandas as pd
%matplotlib inline


#Suavizado exponencial
def ES(alpha,y_0,x):
    y= np.empty(len(x), float)
    y[0]=y_0
    for i in range (1, len(x)):
        y[i]=y[i-1]+alpha*(x[i-1]-y[i-1])
    return y.copy()
data =[110,134,162,180,118,109,114,173,154,132]
pred_esopt=ES(0.5,110,data)
df=pd.DataFrame()
df['data'] = data
df['se_opt']= pred_esopt

df
```

```python
#Gráfico rápido
df[['data','se_opt']].plot(figsize=(16,6))
```

```python
#MAD
def MADn(data,pred,n):
    aux=0
    j=0
    while np.isnan(pred[j]):
        j+=1
    for i in range (j,n+j): 
        aux+=abs(data[i]-pred[i])
    resultado=aux/n
    return resultado

MADn(data,pred_esopt,10)
```

```python
#TS
def TSn(data,pred,n):
    aux=0
    aux2=0
    j=0
    while np.isnan(pred[j]):
        j+=1
    for i in range(j,n+j):
        aux+= data[i]-pred[i]
        aux2+= abs(data[i] - pred[i])
    resultado=n*(aux/aux2)
    return resultado
    
TSn(data,pred_esopt,10)
```
